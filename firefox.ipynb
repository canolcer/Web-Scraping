```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from bs4 import BeautifulSoup
import time

# Provide the path to the GeckoDriver executable (Firefox driver)
driver_path = "path/to/geckodriver"

# URL to scrape
url = 'https://---' #url goes here
# Set up Firefox options
firefox_options = Options()
firefox_options.headless = True  # To run Firefox in headless mode (without opening a window)

# Initialize the FirefoxDriver with options
driver = webdriver.Firefox(service=Service(driver_path), options=firefox_options)

# Open the URL in the Firefox browser
driver.get(url)

# Scroll down the page to load all comments (you may need to adjust this depending on the page content)
scroll_pause_time = 2
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_pause_time)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# Get the page source after scrolling to load all comments
page_source = driver.page_source

# Close the browser
driver.quit()

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(page_source, "html.parser")

# Find the reviews wrapper
reviews_wrapper = soup.find("div", class_="reviews-wrapper")

# Find all comments
comments = reviews_wrapper.find_all("div", class_="comment")
_2txt = open('2.txt', 'w', encoding='utf-8')
# Extract and print the comments
for comment in comments:
    comment_text = comment.find("div", class_="comment-text").text.strip()
    _2txt.write('yorum:' + comment_text + '\n\n')
    print("Yorum:", comment_text)
_2txt.close()
```
